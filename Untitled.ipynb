```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
#Building the encoder block

class EncoderBlock(nn.Module):
    def __init__(self, in_channels, out_channels):
        super().__init__()
        self.block = nn.Sequential(
            nn.Conv2d(in_channels, out_channels, kernel_size=3, padding=1),
            nn.BatchNorm2d(out_channels),
            nn.ReLU(inplace=True),
            nn.Conv2d(out_channels, out_channels, kernel_size=3, padding=1),
            nn.BatchNorm2d(out_channels),
            nn.ReLU(inplace=True)
        )

    def forward(self, x):
        return self.block(x)
```

```python
#Building the decoder block

class DecoderBlock(nn.Module):
    def __init__(self, in_channels, skip_channels, out_channels):
        super().__init__()
        self.block = nn.Sequential(
            nn.Conv2d(in_channels + skip_channels, out_channels, kernel_size=3, padding=1),
            nn.BatchNorm2d(out_channels),
            nn.ReLU(inplace=True),
            nn.Conv2d(out_channels, out_channels, kernel_size=3, padding=1),
            nn.BatchNorm2d(out_channels),
            nn.ReLU(inplace=True)
        )

    def forward(self, x, skip):
        # Upsample to double the spatial size
        x = F.interpolate(x, scale_factor=2, mode='bilinear', align_corners=False)
        # Concatenate the skip connection from the encoder
        x = torch.cat([x, skip], dim=1)
        return self.block(x)
```

```python
#Assembling the Full Network

class DepthEstimationUNet(nn.Module):
    def __init__(self):
        super().__init__()

        # Encoder
        self.enc1 = EncoderBlock(3, 64)    # RGB input → 64 channels
        self.enc2 = EncoderBlock(64, 128)  # 64 → 128 channels (deeper features)
        self.pool = nn.MaxPool2d(kernel_size=2, stride=2)

        # Bottleneck
        self.bottleneck = EncoderBlock(128, 256)  # smallest spatial size, richest features

        # Decoder
        self.dec2 = DecoderBlock(256, 128, 128)  # bottleneck + skip2
        self.dec1 = DecoderBlock(128, 64, 64)    # dec2 output + skip1

        # Final output: 1×1 Conv → single channel depth map
        self.output_conv = nn.Conv2d(64, 1, kernel_size=1)

    def forward(self, x):
        # --- Encoder ---
        skip1 = self.enc1(x)          # Save for skip connection 1
        x = self.pool(skip1)          # Downsample

        skip2 = self.enc2(x)          # Save for skip connection 2
        x = self.pool(skip2)          # Downsample

        # --- Bottleneck ---
        x = self.bottleneck(x)        # Deepest features

        # --- Decoder ---
        x = self.dec2(x, skip2)       # Upsample + merge with skip2
        x = self.dec1(x, skip1)       # Upsample + merge with skip1

        # --- Output ---
        depth = self.output_conv(x)   # 1×H×W depth map
        depth = torch.sigmoid(depth)  # Normalize to [0, 1]

        return depth
```

```python
#Custom dataset class

import os
from PIL import Image
import numpy as np
from torch.utils.data import Dataset, DataLoader
import torchvision.transforms as transforms

class NYUDepthDataset(Dataset):
    def __init__(self, root_dir, split='train', img_size=(256, 256)):
        """
        root_dir : path to nyu_train_preprocessed OR nyu_test_preprocessed
        split    : just for display/debugging purposes
        img_size : resize everything to this (H, W)
        """
        self.img_dir   = os.path.join(root_dir, 'images')
        self.depth_dir = os.path.join(root_dir, 'depth')
        self.img_size  = img_size

        # Collect all filenames (sorted so image[i] matches depth[i])
        self.filenames = sorted(os.listdir(self.img_dir))
        print(f"[{split}] Found {len(self.filenames)} samples")

        # RGB image transform
        self.img_transform = transforms.Compose([
            transforms.Resize(img_size),
            transforms.ToTensor(),                        # [0,255] uint8 → [0,1] float32
            transforms.Normalize(mean=[0.485, 0.456, 0.406],  # ImageNet mean/std
                                 std=[0.229, 0.224, 0.225])
        ])

    def __len__(self):
        return len(self.filenames)

    def __getitem__(self, idx):
        name = self.filenames[idx]                        # e.g. "000000.jpg"
        stem = os.path.splitext(name)[0]                 # "000000"

        # --- Load RGB image ---
        img_path = os.path.join(self.img_dir, name)
        image = Image.open(img_path).convert('RGB')
        image = self.img_transform(image)                # Tensor [3, H, W]

        # --- Load depth map ---
        depth_path = os.path.join(self.depth_dir, stem + '.png')
        depth = Image.open(depth_path)                   # grayscale or 16-bit PNG
        depth = depth.resize(self.img_size, Image.NEAREST)
        depth = np.array(depth, dtype=np.float32)

        # Normalize depth to [0, 1]
        depth_max = depth.max()
        if depth_max > 0:
            depth = depth / depth_max

        depth = torch.from_numpy(depth).unsqueeze(0)    # [1, H, W]

        return image, depth
```

```python
#Loss function

import torch
import torch.nn.functional as F

def ssim_loss(pred, target, window_size=11):
    """
    Structural Similarity loss — penalizes structural differences,
    not just pixel-level errors. Ranges from 0 (identical) to 1 (completely different).
    """
    C1 = 0.01 ** 2
    C2 = 0.03 ** 2

    mu_p = F.avg_pool2d(pred,   window_size, 1, window_size // 2)
    mu_t = F.avg_pool2d(target, window_size, 1, window_size // 2)

    mu_p_sq = mu_p ** 2
    mu_t_sq = mu_t ** 2
    mu_pt   = mu_p * mu_t

    sigma_p  = F.avg_pool2d(pred ** 2,        window_size, 1, window_size // 2) - mu_p_sq
    sigma_t  = F.avg_pool2d(target ** 2,      window_size, 1, window_size // 2) - mu_t_sq
    sigma_pt = F.avg_pool2d(pred * target,    window_size, 1, window_size // 2) - mu_pt

    ssim = ((2 * mu_pt + C1) * (2 * sigma_pt + C2)) / \
           ((mu_p_sq + mu_t_sq + C1) * (sigma_p + sigma_t + C2))

    return torch.clamp((1 - ssim) / 2, 0, 1).mean()


def depth_loss(pred, target, alpha=0.85):
    """
    Combined loss:
      alpha     * SSIM  → preserves structure (edges, object boundaries)
      (1-alpha) * L1    → ensures absolute depth values are correct
    """
    l1   = F.l1_loss(pred, target)
    ssim = ssim_loss(pred, target)
    return alpha * ssim + (1 - alpha) * l1
```

```python
#training loop

import torch
from torch.optim import Adam
from torch.optim.lr_scheduler import ReduceLROnPlateau

def train_model(
    train_dir,
    test_dir,
    num_epochs   = 30,
    batch_size   = 8,
    lr           = 1e-4,
    img_size     = (256, 256),
    save_path    = "best_model.pth"
):
    # ── Device ──────────────────────────────────────────────────────────────
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    print(f"Training on: {device}")

    # ── Datasets & DataLoaders ───────────────────────────────────────────────
    train_dataset = NYUDepthDataset(train_dir, split='train', img_size=img_size)
    test_dataset  = NYUDepthDataset(test_dir,  split='test',  img_size=img_size)

    train_loader = DataLoader(train_dataset, batch_size=batch_size,
                              shuffle=True,  num_workers=2, pin_memory=True)
    test_loader  = DataLoader(test_dataset,  batch_size=batch_size,
                              shuffle=False, num_workers=2, pin_memory=True)

    # ── Model, Optimizer, Scheduler ─────────────────────────────────────────
    model     = DepthEstimationUNet().to(device)
    optimizer = Adam(model.parameters(), lr=lr)
    scheduler = ReduceLROnPlateau(optimizer, mode='min', patience=3,
                                  factor=0.5, verbose=True)

    best_val_loss = float('inf')
    history = {'train_loss': [], 'val_loss': []}

    # ── Epoch Loop ───────────────────────────────────────────────────────────
    for epoch in range(num_epochs):

        # ---- Training phase ------------------------------------------------
        model.train()
        train_loss = 0.0

        for batch_idx, (images, depths) in enumerate(train_loader):
            images = images.to(device)
            depths = depths.to(device)

            optimizer.zero_grad()          # Clear old gradients
            preds = model(images)          # Forward pass
            loss  = depth_loss(preds, depths)  # Compute loss
            loss.backward()                # Backprop: compute gradients
            optimizer.step()               # Update weights

            train_loss += loss.item()

            # Print progress every 20 batches
            if (batch_idx + 1) % 20 == 0:
                print(f"  Epoch [{epoch+1}/{num_epochs}] "
                      f"Batch [{batch_idx+1}/{len(train_loader)}] "
                      f"Loss: {loss.item():.4f}")

        avg_train_loss = train_loss / len(train_loader)

        # ---- Validation phase ----------------------------------------------
        model.eval()
        val_loss = 0.0

        with torch.no_grad():              # No gradients needed for validation
            for images, depths in test_loader:
                images = images.to(device)
                depths = depths.to(device)
                preds  = model(images)
                val_loss += depth_loss(preds, depths).item()

        avg_val_loss = val_loss / len(test_loader)
        scheduler.step(avg_val_loss)       # Adjust LR if val loss plateaus

        # ---- Logging -------------------------------------------------------
        history['train_loss'].append(avg_train_loss)
        history['val_loss'].append(avg_val_loss)

        print(f"\nEpoch [{epoch+1}/{num_epochs}] "
              f"Train Loss: {avg_train_loss:.4f} | "
              f"Val Loss: {avg_val_loss:.4f}")

        # ---- Save best model -----------------------------------------------
        if avg_val_loss < best_val_loss:
            best_val_loss = avg_val_loss
            torch.save({
                'epoch':      epoch + 1,
                'model_state_dict':     model.state_dict(),
                'optimizer_state_dict': optimizer.state_dict(),
                'val_loss':   best_val_loss,
            }, save_path)
            print(f"  ✓ Best model saved (val_loss = {best_val_loss:.4f})")

        print("-" * 60)

    print(f"\nTraining complete. Best val loss: {best_val_loss:.4f}")
    return model, history
```
