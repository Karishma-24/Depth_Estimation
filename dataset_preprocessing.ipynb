Preprocessing training data

```python
#source root
#source_root = r"C:\Users\KARISHMA\Desktop\deep learning\Depth_Estimation\dataset\data\nyu2_train"
from pathlib import Path

source_root = Path("dataset/data/nyu2_train")
```

```python
#output root
output_root = "dataset/preprocessed_data/nyu_train_preprocessed"
images_out = os.path.join(output_root, "images")
depth_out = os.path.join(output_root, "depth")

#making output files
os.makedirs(images_out, exist_ok=True)
os.makedirs(depth_out, exist_ok=True)
```

```python
#Segregation code

counter = 0

for scene in os.listdir(source_root):
    scene_path = os.path.join(source_root, scene)

    # Skip if not a folder
    if not os.path.isdir(scene_path):
        continue

    files = os.listdir(scene_path)

    jpg_files = [f for f in files if f.endswith(".jpg")]

    for jpg in jpg_files:
        base_name = os.path.splitext(jpg)[0]
        png_name = base_name + ".png"

        if png_name in files:
            rgb_path = os.path.join(scene_path, jpg)
            depth_path = os.path.join(scene_path, png_name)

            new_name = f"{counter:06d}"

            shutil.copy(rgb_path, os.path.join(images_out, new_name + ".jpg"))
            shutil.copy(depth_path, os.path.join(depth_out, new_name + ".png"))

            counter += 1

print(f"Done! Total samples: {counter}")
```
