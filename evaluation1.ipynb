```python
#imports
import os
import numpy as np
import cv2
```

```python
#image paths
pred_path = "dataset/nyu_test_processed/images" #rgb image
gt_path = "dataset/nyu_test_processed/depth" #depth image

pred_files = sorted([f for f in os.listdir(pred_path) if f.endswith(".npy")])
```

```python
#Absolute Relative Error list (measures relative error)
abs_rel_list = []

#RMSE list (penalizes large errors)
rmse_list = []
```

```python
#evaluation and printing results of absolute relative error and RMSE

for file in pred_files:
    base = file.replace("_disp.npy", "")
    
    pred_disp = np.load(os.path.join(pred_path, file))
    pred_depth = 1.0 / (pred_disp + 1e-6)

    gt_file = base + ".png"
    gt_depth = cv2.imread(os.path.join(gt_path, gt_file), cv2.IMREAD_UNCHANGED).astype(np.float32)

    #Resizing prediction to GT size
    pred_depth = cv2.resize(pred_depth, (gt_depth.shape[1], gt_depth.shape[0]))

    #Normalize GT depth 
    gt_depth = gt_depth / 1000.0 #mm -> meters

    mask = gt_depth > 0
    
    pred = pred_depth[mask]
    gt = gt_depth[mask]

    #Metrics
    abs_rel = np.mean(np.abs(gt - pred) / gt)
    rmse = np.sqrt(np.mean((gt - pred) ** 2))

    abs_rel_list.append(abs_rel)
    rsme_list.append(rmse)

#printing results
print("Final Results:")
print("Absolute Relative Error:", np.mean(abs_rel_list))
print("RMSE:", np.mean(rmse_list))
```
